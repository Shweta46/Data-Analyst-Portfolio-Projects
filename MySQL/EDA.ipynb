### Seeing, lets consider a test case, that how much a particular continent grew and how fast 

```python
# Select only numeric columns
nc = df.select_dtypes(include=['float64', 'int64'])

# Group by 'Continent' and calculate the mean of numeric columns
mean_by_continent = nc.groupby(df['Continent']).mean().sort_values(by='2022 Population', ascending=False)
mean_by_continent
```

```python
df[df['Continent'].str.contains('Oceania')]
```

```python
# finding index of columns that we want to actually see the plot of
df.columns
# Columns from index number 5 to 13 is the one
```

```python
# only selecting the columns that we want to see the graph of 
df2 = nc.groupby(df['Continent'])[df.columns[5:13]].mean().sort_values(by='2022 Population', ascending=False)
df2
```

```python
df3 = df2.transpose()
df3
```

```python
df3.plot()
```

### But we want the display to be from 1970 to 2020. So, we need to actually specify the columns in the df2 instead of just writing the index numbers

```python
df2 = nc.groupby(df['Continent'])[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean().sort_values(by='2022 Population', ascending=False)
df2
```

```python
df3 = df2.transpose()
df3
```

```python
df3.plot()
```
