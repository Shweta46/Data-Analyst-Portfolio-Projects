```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv("top50.csv", encoding= "ISO-8859-1", index_col=[0])
```

```python
df.columns = df.columns.str.strip()
```

```python
df.columns = df.columns.str.replace(".", "")
```

```python
df.head()
```

```python
conn = sqlite3.connect("database.db")
c = conn.cursor()
```

```python
df.to_sql("TopSongss", conn)
```

```python
c.execute('''select *
from TopSongss;''')
data = pd.DataFrame(c.fetchall())
data.columns = [x[0] for x in c.description]
data
```

```python
c.execute('''select track_name, energy,
(case when energy > 0.66 then 'above average'
when energy < 0.66 then 'below average'
when energy = 0.66 then 'Average' end) as ComparingAverage
from TopSongss
limit 10
;''')
data = pd.DataFrame(c.fetchall())
data.columns = [x[0] for x in c.description]
data
```

```python
c.execute('''select track_name, energy,
(case 
    when energy between 0.54 and 0.74 then 'Within 10'
    when (energy between 0.44 and 0.54) or (energy between 0.74 and 0.84) then 'Within 20'
    when (energy between 0.34 and 0.44) or (energy between 0.84 and 0.94) then 'within 30'
    else 'greater than 0.3'
end) as CompareAverage
from TopSongss
;''')
data = pd.DataFrame(c.fetchall())
data.columns = [x[0] for x in c.description]
data
```

```python
c.execute('''
select 
    sum(case
        when energy between 0.54 and 0.74 then 1
        else 0
    end) as within10,
    sum(case 
        when (energy between 0.44 and 0.54) or (energy between 0.74 and 0.84) then 1
        else 0
    end) as within20,
    sum(case
        when (energy between 0.34 and 0.44) or (energy between 0.84 and 0.94) then 1
        else 0
    end) as within30,
    sum(case
        when (energy < 0.34) or (energy > 0.94) then 1
        else 0        
    end) as greater30
from TopSongss;''')
data = pd.DataFrame(c.fetchall())
data.columns = [x[0] for x in c.description]
data
```
